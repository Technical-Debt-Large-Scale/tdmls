### 1.2.2 Gráfico de Correlação entre Technical Debt x LeadTime

```python
arrayLeadTime = df_all_metrics.leadTime.values
```

```python
arrayTechnicalDebt = df_all_metrics.technicalDebt.values
```

```python
# plot
pyplot.scatter(arrayLeadTime, arrayTechnicalDebt)
pyplot.show()
```

### 1.2.2.1 Calculate spearman's correlation (LeadTime x TD)

```python
# calculate spearman's correlation
coef1, p1 = stats.spearmanr(arrayLeadTime, arrayTechnicalDebt)
```

```python
print('Spearmans correlation coefficient: %.3f' % coef1)
# interpret the significance
alpha = 0.05
if p1 > alpha:
	print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p1)
else:
	print('Samples are correlated (reject H0) p=%.3f' % p1)
```

#### LeadTime e TD são correlacionados (OK!)

```python
sns.lmplot(x='leadTime', y='technicalDebt', data=df_all_metrics)
```
