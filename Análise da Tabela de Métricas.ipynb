### 1.3.1 Gráfico de Correlação entre LeadTime e Technical Debt

```python
arrayLeadTime = df_all_metrics.leadTime.values
```

```python
arrayTechnicalDebt = df_all_metrics.technicalDebt.values
```

```python
# plot
pyplot.scatter(arrayLeadTime, arrayTechnicalDebt)
pyplot.show()
```

```python
sns.lmplot(x='leadTime', y='technicalDebt', data=df_all_metrics)
```

### 1.3.1.1 Calculate spearman's correlation (LeadTime x TD)

```python
stats.spearmanr(arrayLeadTime, arrayTechnicalDebt)
```

```python
type(stats.spearmanr(arrayLeadTime, arrayTechnicalDebt))
```

```python
# calculate spearman's correlation
coefLeadTime, pLeadTime = stats.spearmanr(arrayLeadTime, arrayTechnicalDebt)
```

```python
print('Spearmans correlation coefficient: %.3f' % coefLeadTime)
# interpret the significance
alpha = 0.05
if pLeadTime > alpha:
	print('Samples are uncorrelated (fail to reject H0) p=%.3f' % pLeadTime)
else:
	print('Samples are correlated (reject H0) p=%.3f' % pLeadTime)
```

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>
    Pela análise do coeficiente de Spearman, pode-se perceber que <u>LeadTime</u> e <u>TD</u> <b>são</b> <u>correlacionados</u>. Com isso, existe uma tendência de que na medida que o LeadTime aumenta a Dívida Técnica tende a aumentar
</p>

## 1.4 Resumo da análise de correlação (Spearman's correlation coefficient)

|  Caracterítica     | Dívida Técnica  | Correlacionado |  
|--------------------|-----------------|----------------|
| LeadTime           | TD              | SIM            |
| ComplexityPoints   | TD              | SIM            |
| TotalDevelopers    | TD              | SIM            |
| TaskScaling        | TD              | SIM            | 
| Maturity           | TD              | Não            | 
| TaskGlobalDistance | TD              | Não            |

```python
from tabulate import tabulate
```

```python
def showCalcPSpearman(p, alfa):
    if p > alfa:
        return 'NÃO'
    else:
        return 'SIM'
```

```python
lineLeadTime = ['LeadTime', 'TD', coefLeadTime, showCalcPSpearman(pLeadTime, 0.05)]
lineComplexityPoints = ['ComplexityPoints', 'TD', coefComplexityPoints, showCalcPSpearman(pComplexityPoints, 0.05)]
lineTotalDevelopers = ['TotalDevelopers', 'TD', coefTotalDevelopers, showCalcPSpearman(pTotalDevelopers, 0.05)]
lineTaskScaling = ['TaskScaling', 'TD', coefTaskScaling, showCalcPSpearman(pTaskScaling, 0.05)]
lineMaturity = ['Maturity', 'TD', coefMaturity, showCalcPSpearman(pMaturity, 0.05)]
lineTaskGlobalDistance = ['TaskGlobalDistance', 'TD', coefTaskGlobalDistance, showCalcPSpearman(pTaskGlobalDistance, 0.05)]
analysisHeaders = ['Característica', 'Dívida Técnica', 'Spearmans Coef', 'Correlacionado']

print(tabulate([lineLeadTime, lineComplexityPoints, lineTotalDevelopers, 
                lineTaskScaling, lineMaturity, 
                lineTaskGlobalDistance],
               headers=analysisHeaders))
```
