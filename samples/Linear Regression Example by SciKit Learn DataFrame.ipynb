### Importando as bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

import pandas as pd
import seaborn as sns
```

### Carrega o dataset diabetes

```python
# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

### Converte diabetes_X e diabetes_y e para DataFrames

```python
X = pd.DataFrame(data=diabetes_X, columns=['x1', 'x2', 'x3','x4', 'x5', 'x6', 'x7', 'x8','x9', 'x10'])
```

```python
X.head(10)
```

```python
df_X_y = X
```

```python
y = pd.DataFrame(data=diabetes_y, columns=['y'])
```

```python
y.head(10)
```

```python
df_X_y['y'] = y
```

```python
df_X_y.head(10)
```

### Plota o gráfico de regressão linear para uma feature

```python
# Plot outputs
plt.scatter(X['x1'].values, y.values,  color='black')
plt.show()
sns.lmplot(x='x1', y='y', data=df_X_y)
```

### Matriz de correlação

```python
df_X_y.corr().round(3)
```

### Pairplot

```python
sns.pairplot(df_X_y)
plt.show()
```

### Importa o train_test_split do scikit learn

```python
from sklearn.model_selection import train_test_split
```

### Criando os dados de treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Cria um objeto de Regressão Linear

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
regr
```

### Treina o modelo usando os dados de treinamento

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

### Cria o modelo de predição usando os dados de testes

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
y_pred.shape
```

### Mostra os coeficientes de regressão, o Mean Squared Error e Coeficiente de determinação

```python
# The coefficients
print('Coefficients: \n', regr.coef_.round(4))
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
```
